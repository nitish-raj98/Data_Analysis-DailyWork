```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
df.columns = df.columns.str.replace("-","_")
```

```python
df.columns
```

```python
labels = ["good", "better", "best"]
df["quantity_segregation"] = pd.qcut(df["quantity"], q=3, labels=labels)
```

```python
df[["quantity", "quantity_segregation"]].head(10)
```

```python
#create a column for purchase frequency
frequency_mapping = {
    "fortnightly" : 140,
    "weekly" : 850,
    "monthly" :300,
    "quarterly" : 900,
    "annually" : 365,
    "every_3_months" :90
}

df["servicing_frequency_days"] = df["sales"].map(frequency_mapping)
```

```python
df.columns
```

```python
df[["servicing_frequency_days", "ship_date"]].head(10)
```

```python
print(df.columns)
```

```python
df = df.drop(["servicing_frequency_days"], axis=1)
```

```python
import numpy as np

df["discount_flag"] = np.where(df["discount"] > 0, "yes-discount", "no-discount")
```
